```python
import matplotlib as plt
import numpy as np
import pandas as pd
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2023", periods = 1000))
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=pd.date_range("1/1/2023", periods = 1000), columns=list("ABCD"))
df
```

```python
df = df.cumsum()
df.plot()
```

```python
import matplotlib.pyplot as plt1
```

```python
ypoints = np.array([3, 8, 1, 10])
plt1.plot(ypoints, marker = 'o')
plt1.show
```

```python
ypoints = np.array([3, 8, 1, 10])
plt1.plot(ypoints, marker = 'o', linestyle = 'dotted')
plt1.show
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt1.title("Sports Watch Data")
plt1.xlabel("Average Pulse")
plt1.ylabel("Calorie Burnage")

plt1.plot(x, y)

plt1.grid()

plt1.show()
```
