```python
import matplotlib as plt
import numpy as np
import pandas as pd
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2023", periods = 1000))
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=pd.date_range("1/1/2023", periods = 1000), columns=list("ABCD"))
df
```

```python
df = df.cumsum()
df.plot()
```
